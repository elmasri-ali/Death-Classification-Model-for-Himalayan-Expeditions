# Introduction

The Himalayas, home to the world's highest peaks, have long been a magnet for mountaineers. With this allure comes inherent risks, including extreme weather, avalanches, and altitude sickness, contributing to a significant number of fatalities. This project aims to build a classification model to predict the probability of death for summiteers during their climb in the Himalayas. 

### Data:

I developed the model using historical data from 1905 to 2020 obtained from Kaggle.com. The data consists of four CSV files; here's a brief description for each one:

* Deaths Dataset: 

Contains information about individuals who died during expeditions, including details like peak name, date, time, citizenship, gender, age, oxygen use, summit status, and cause of death.

* Expeditions Dataset:

Provides data on various expeditions, including peak name, nationality of the expedition, year, season, sponsors, leaders, and other details related to the expedition.

* Peaks Dataset: 

Lists of different peaks, their height, whether they have been climbed, and information about the first ascent.

* Summiters Dataset:

Details individuals who summited, including their name, season of the climb, date, time, citizenship, gender, age, oxygen use, and if they died on the descent.



Data source link: https://www.kaggle.com/datasets/raskoshik/himalayan-expeditions/data

### Objective: 

Despite advancements in climbing technology and weather forecasting, the Himalayas continue to pose severe risks to climbers and claim numerous lives yearly. Understanding the factors contributing to these fatalities can help plan safer expeditions and save lives. The primary problem is identifying the key determinants that increase the likelihood of death during a climb and eventually training a model that can predict the outcome variable. The target variable for the classification model will be a binary outcome indicating whether a climber died during the expedition. The model should utilize various features such as age, gender, experience, peak difficulty, season of climb, and weather conditions to predict survival chance for potential future climbers.

## Season Mapping 

```python
# Define a mapping from season abbreviation to a numerical value
season_mapping = {
    'Aut': 1,  # Autumn
    'Win': 2,  # Winter
    'Sum': 3,  # Summer
    'Spr': 4   # Spring
}

# Define a function to map the season abbreviation to a number
def map_season_to_number(yr_season_str):
    # Extract the season abbreviation (last three characters)
    season_abbr = yr_season_str[-3:]
    # Map abbreviation to number
    return season_mapping.get(season_abbr, 0)  # Return 0 if the season is not found

# Apply the function to the 'yr_season' column to create a new 'season_number' column
summiters_df['season_number'] = summiters_df['yr_season'].apply(map_season_to_number)
deaths_df['season_number'] = deaths_df['yr_season'].apply(map_season_to_number)
# Display the first few rows to verify the new 'season_number' column
summiters_df[['yr_season', 'season_number']].head()
```

### Engineering a "Climb Count" feature for each climber.

```python
# Add a 'Climb Count' feature for each climber
combined_df['Climb Count'] = combined_df.groupby('name')['name'].transform('count')

# Calculate the number of climbs per peak as 'Peak Count'
combined_df['Peak Count'] = combined_df.groupby('peak_id')['peak_id'].transform('count')

# Drop the previously decided irrelevant features
df_revised = combined_df.drop(['peak_name', 'host_country', 'full_date', 
                              'cause_of_death', 'is_summiter'], axis=1)
```

### Convert to Binary

```python
# For the initial preprocessing steps, convert 'is_o2_used' and 'died_on_descent' to binary
df_revised['is_o2_used'] = df_revised['is_o2_used'].map({'Yes': 1, 'No': 0}).fillna(0)  # Assuming non-use for missing
df_revised['died_on_descent'] = df_revised['died_on_descent'].apply(lambda x: 0 if x == '.' else 1) # Mapping . to 0
df_revised['gender'] = df_revised['gender'].map({'M': 1, 'F': 0}) # Converting gender 


df_revised.head()
```

### Imputing Missing Values and Citizenships Assessment

```python
# Assessing missing data for new and retained features
missing_data_final = df_revised.isnull().sum()

#let's assess the number of unique categories to ensure feasibility
unique_citizenships = df_revised['citizenship'].nunique()

missing_data_final, unique_citizenships
```

```python
# Imputing missing values for 'height_m' and 'climb_status'
df_revised['height_m'].fillna(df_revised['height_m'].median(), inplace=True)
df_revised['climb_status'].fillna(df_revised['climb_status'].mode()[0], inplace=True)

# Handling 'citizenship' by grouping less frequent categories into 'Other'
# Determine the threshold for grouping
citizenship_counts = df_revised['citizenship'].value_counts()
threshold = 100 
frequent_citizenships = citizenship_counts[citizenship_counts > threshold].index.tolist()

# Grouping less frequent citizenships into 'Other'
df_revised['citizenship_grouped'] = df_revised['citizenship'].apply(lambda x: x if x in frequent_citizenships else 'Other')

# Dropping original 'citizenship' to avoid duplication
df_revised.drop('citizenship', axis=1, inplace=True)

# Final check before one-hot encoding for 'citizenship_grouped' and potentially 'climb_status'
df_revised.head()
```

### Model Selection and Evaluation

In this phase of the project, my goal is to construct a systematic approach for evaluating a range of classification models to determine the most suitable one for our specific dataset, which is marked by a binary outcome variable alongside predominantly binary features. The search for the optimal model is challenged by the notable imbalance within both the features and the outcome variable of the dataset, a critical factor that necessitates careful consideration.
To effectively navigate this challenge, my strategy encompasses training models on both the original dataset and a version where the dataset is resampled to correct for its inherent imbalance. This dual approach ensures a comprehensive assessment of each model's performance under different conditions. Furthermore, I explored various hyperparameters across models to unearth the configurations that yield the highest efficacy, considering both scenarios with and without the application of Principal Component Analysis (PCA) for dimensionality reduction.
Evaluating model performance in this context requires a nuanced set of metrics due to the dataset's imbalanced nature and the binary classification task at hand. Therefore, I prioritized metrics that offer a more detailed view of a model's ability to distinguish between classes under these conditions, such as precision, recall, and the precision-recall Area Under the Curve (AUC). These metrics are especially pertinent as they provide a deeper insight into model performance across both the original dataset and its resampled counterpart, ensuring a holistic understanding of each model's capabilities in managing the dataset's unique challenges.


```python
# Pretty print Evaluation Function 
def evaluate_model(y_true, y_pred, y_proba, model_name="Model"):
    """
    Evaluate the model and print out the performance metrics.

    Parameters:
    - y_true: Array-like of test target values
    - y_pred: Array-like of predicted target values
    - y_proba: Array-like of predicted probabilities for the positive class
    - model_name: String, name of the model

    Returns:
    - None, prints the evaluation metrics.
    """
    # Calculate metrics
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_proba)
    
    # Calculate Precision-Recall AUC
    precision_curve, recall_curve, _ = precision_recall_curve(y_true, y_proba)
    auc_pr = auc(recall_curve, precision_curve)
    
    # Print metrics
    print(f"{model_name} Evaluation Metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC-AUC: {roc_auc:.4f}")
    print(f"Precision-Recall AUC: {auc_pr:.4f}")
    print("-"*30)  
```

```python
# Best estimator prediction function 
def predict_with_best_model(gridsearch, X_test):
    """
    Use the best estimator from the grid search to make predictions and predict probabilities.

    Parameters:
    - gridsearch: The GridSearchCV object after fitting
    - X_test: The test set features

    Returns:
    - y_pred: Predicted classes
    - y_pred_proba: Predicted probabilities for the positive class
    """
    best_model = gridsearch.best_estimator_
    y_pred = best_model.predict(X_test)
    y_pred_proba = best_model.predict_proba(X_test)[:, 1]
    return y_pred, y_pred_proba
```

```python
# Fit the grid search on the training data
gridsearch_no_pca = GridSearchCV(
    pipe_no_pca,
    search_space_no_pca,
    cv=5,
    verbose=1,
    scoring='roc_auc',  #  ROC-AUC for imbalanced datasets
    refit='roc_auc',  # Ensure refitting is done on the most informative metric
    n_jobs=-1  # Utilize all available cores for efficiency
)
%time gridsearch_no_pca.fit(X_train, y_train)  # Fit on training data
```

```python
y_pred_no_pca, y_pred_proba_no_pca = predict_with_best_model(gridsearch_no_pca, X_test)
```

```python
evaluate_model(y_test, y_pred_no_pca, y_pred_proba_no_pca, model_name="Non-PCA Without Resampling")
```

### Conclusion

Based on the detailed evaluation of few models, the best-performing model (**random_search_no_pca_re**) was determined through a comprehensive analysis of several metrics, with a particular focus on the **PR-Area Under the Curve (PR-AUC) of 0.7809 and Precision of Precision: 0.9661 ** scores instead of just accuracy and ROC. This choice of metric is especially pertinent in the context of imbalanced datasets, like the one at hand, where the positive class (deaths in Himalayan expeditions) is significantly less frequent and sensitive than the negative one.

The evaluation encompassed models trained both with and without Principal Component Analysis (PCA) and considered the impact of resampling the dataset to address class imbalance. The models evaluated included K-Nearest Neighbors, Logistic Regression, Random Forest, Support Vector Machines, and Decision Trees, each with various hyperparameters. The evaluations were done in two scenarios: using the original dataset and a resampled dataset using the Synthetic Minority Over-sampling Technique (SMOTE) to counteract the imbalance issue.

**The results demonstrated that the RandomForestClassifier, particularly when configured with balanced class weights, a max depth of 20, min_samples_split of 10, and n_estimators set to 500, achieved superior performance metrics**. This configuration was chosen as the best model in both non-PCA and PCA scenarios, with slight variations in hyperparameters between the two. The model's robustness was evident from its high PR-AUC scores, which are indicative of its ability to distinguish between the positive and negative classes effectively. The PR-AUC score is a critical metric in this analysis because it provides a comprehensive measure of performance across all classification thresholds, unlike accuracy, which can be misleading in imbalanced datasets.

**In the PCA scenario with and without resampling, the model's performance decreased.**  The
 application of PCA did not result in better performance compared to the non-PCA models.These scores, while respectable, were lower than those achieved by the  non-PCA models, suggesting that dimensionality reduction might not be necessary or beneficial fo 
 this particular dataset and task 

The conclusion drawn frour3's evaluations is that the RandomForestClassifier without PCA, particularly with class weights balanced and after resampling the dataset, fits the model best. This conclusion is supported by its high precision, recall, F1 score, ROC-AUC and especially its PR-AUC score, which indicates a strong ability to distinguish between the class labels in an imbalanced dataset. The choice of PR-AUC over accuracy as the primary metric for model evaluation ensures that the model's performance is assessed more holistically, acknowledging the dataset's imbalanced nature.
The doubt over the model being overfitted or underfitted is mitigated by the use of cross-validation and grid search techniques, which help in selecting the best hyperparameters while ensuring the model's generalizability across different subsets of the data. The consistent performance of the model across different metrics, including precision, recall, F1 score, and ROC-AUC, further supports its robustness and suitability for predicting deaths in Himalayan expeditions with a high degree of reliability.
